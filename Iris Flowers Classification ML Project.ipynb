# Iris Flowers Classification ML Project

This particular ML project is usually referred to as the “Hello World” of Machine Learning. The iris flowers dataset contains numeric attributes,

Import Requires LIbraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Import Data

```python
columns= ['Sepal_length(cm)', 'Sepal_Width(cm)','Petal_length(cm)','Petal_Width(cm)', 'Class:']
df = pd.read_csv('iris.data', names=columns)
df.head()
```

```python
df.describe()
```

Visualizations:

```python
sns.pairplot(df, hue="Class:", diag_kind="hist",markers=['o','s','D'])
```

```python
df.isnull().sum()
```

```python
df['Class:'].value_counts()
```

```python
sns.countplot(df['Class:']);
```

```python
X=np.asarray(df[['Sepal_length(cm)', 'Sepal_Width(cm)','Petal_length(cm)','Petal_Width(cm)']])
X[0:5]
```

```python
y=np.asarray(df[['Class:']])
y[0:5]
```

Split Data into Train and Test Datasets

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,random_state =0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Modeling

```python
from sklearn.linear_model  import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

In a Jupyter environment, please rerun this cell to show the HTML representation or trust the notebook.
On GitHub, the HTML representation is unable to render, please try loading this page with nbviewer.org.

```python
yhat=model.predict(X_test)
```

```python
yhat
```

```python
confusion_matrix(y_test,yhat)
```

```python
accuracy=accuracy_score(y_test,yhat)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```
